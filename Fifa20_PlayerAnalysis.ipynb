# **CREAR SESSION IN SPARK**

```python
from pyspark.sql import SparkSession
from pyspark import SparkContext

SpSession = SparkSession \
          .builder \
          .appName("Fifa20_Analysis_with_Spark") \
          .getOrCreate()
```

```python
path="/content/fifa_csv/players_20.csv"
```

```python
data = SpSession.read.csv(path, header=True, sep=",")
data.show(5)
```

```python
data.dtypes
```

```python
data.printSchema()
```

```python
data.columns
```

```python
from os import truncate
#METODO 1

data.select(data.long_name).show(truncate=False)
```

```python
#Método 2 de la selección multiple de columnas

data.select(data['short_name'], data['long_name']).show(truncate=False)
```

# **AGRUPACIÓN**

```python
data.groupBy('nationality').count().show(10)
```

```python
data.groupBy('nationality','player_positions').count().show(10)
```

# **ORDENAR FILAS**

```python
data.orderBy('long_name').show(truncate=False)
```

```python
data.orderBy('nationality', ascending=False).show(truncate=False)
```

```python
data.groupBy('nationality').count().orderBy('count', ascending=False).show(10)
```

# **FILTRADO DE DATOS**

```python
from pyspark.sql.functions import col

total_count = data.count()
print("Total de registros:", total_count)

num_peruvian_players = data.filter(col('nationality')=='Peru').count()
print("Total peruvian players: ", num_peruvian_players)

data.filter(col('nationality')=='Peru').show(truncate=False)
```

```python
data.filter((col('nationality')=='Peru')&(col('age')<25)).show(truncate=False)
```

```python
data.filter((col('nationality')=='Peru')&(col('age')>28)&(col('international_reputation')>=2)).orderBy('age', ascending=False).show(truncate=False)
```

```python
data.filter((col('nationality')=='Spain')&(col('age')>28)).orderBy('age', ascending=False).show(truncate=False)
```

```python
data.filter((col('age')>24)&(col('international_reputation')>4)).orderBy('age', ascending=False).show(truncate=False)
```
